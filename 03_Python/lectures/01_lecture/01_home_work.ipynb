#### Task 2: Apply a 15% discount

```python
original_price = 120
discount_rate = 0.15  # 15%

discount_amount = original_price * discount_rate
final_price = original_price - discount_amount

print("Original price:", original_price, "EUR")
print("Discount amount:", discount_amount, "EUR")
print("Final price after discount:", final_price, "EUR")
```

#### Task 3: Monthly salary

```python
hours_per_month = 160
hourly_rate = 12  # EUR per hour

monthly_salary = hours_per_month * hourly_rate

print("Monthly salary:", monthly_salary, "EUR")
```

#### Task 4: Profit calculation

```python
cost_price = 40
selling_price = 65

profit = selling_price - cost_price

print("Profit:", profit, "EUR")
```

#### Task 5: Price without VAT

```python
price_with_vat = 121
vat_rate = 0.21  # 21%

price_without_vat = price_with_vat / (1 + vat_rate)

print("Price without VAT:", price_without_vat, "EUR")
```

#### Task 6: Profit after 1 year (5% interest)

```python
principal = 1000
interest_rate = 0.05  # 5%

profit_after_year = principal * interest_rate
total_after_year = principal + profit_after_year

print("Profit after 1 year:", profit_after_year, "EUR")
print("Total after 1 year:", total_after_year, "EUR")
```

#### Task 7: Average of grades

#### Task 8: Unpaid loan percentage

```python
loan_total = 1000
paid_amount = 350

unpaid_amount = loan_total - paid_amount
unpaid_percentage = (unpaid_amount / loan_total) * 100

print("Unpaid amount:", unpaid_amount, "EUR")
print("Unpaid percentage:", unpaid_percentage, "%")
```
