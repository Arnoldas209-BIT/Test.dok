#### Task 1: Compare two numbers a and b

```python
a = 5
b = 10

if a < b:
    print("a is smaller than b")
elif a == b:
    print("a is equal to b")
else:
    print("a is greater than b")
```

#### Task 2: String operations with "Zen of Python"

```python
text = "Zen of Python"

words = text.split()  # ['Zen', 'of', 'Python']

# 1) Print the last character of the second word
print("Last char of 2nd word:", words[1][-1])

# 2) Print the first character of the third word
print("First char of 3rd word:", words[2][0])

# 3) Print only the first word
print("First word:", words[0])

# 4) Print only the last word
print("Last word:", words[-1])

# 5) Print the whole phrase reversed
print("Reversed text:", text[::-1])

# 6) Split words and print them
print("Words:", words)

# 7) Replace "Python" with "Programming" and print new sentence
new_text = text.replace("Python", "Programming")
print("Replaced sentence:", new_text)
```

#### Task 3: Simple calculator

```python
a = 10
b = 5
operation = "*"

if operation == "+":
    print("Result:", a + b)
elif operation == "-":
    print("Result:", a - b)
elif operation == "*":
    print("Result:", a * b)
elif operation == "/":
    if b == 0:
        print("Error: division by zero is not allowed")
    else:
        print("Result:", a / b)
else:
    print("Error: unknown operation")
```

#### Task: Test more string methods

```python
text = "Zen of Python"

print("Original:", text)
print("-" * 30)

print("upper():", text.upper())
print("casefold():", text.casefold())
print("capitalize():", text.capitalize())

print("count('o'):", text.count("o"))
print("count('Python'):", text.count("Python"))

print("find('of'):", text.find("of"))
print("find('Python'):", text.find("Python"))
print("find('Java'):", text.find("Java"))  # -1 means not found

print("replace('Python', 'Programming'):", text.replace("Python", "Programming"))

print("startswith('Zen'):", text.startswith("Zen"))
print("endswith('Python'):", text.endswith("Python"))

print("split():", text.split())
print("join():", " | ".join(text.split()))

print("title():", text.title())
print("swapcase():", text.swapcase())
print("lower():", text.lower())

print("isupper():", text.isupper())
print("islower():", text.islower())
print("isalpha():", text.isalpha())  # False because spaces exist
print("isdigit():", text.isdigit())
```
